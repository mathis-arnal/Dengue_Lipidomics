# Prepare dataset for LipidSig

```python
df = pd.read_csv(
    '/Users/loictalignani/research/project/lipidomics/data/lipidsig/MISSE_Global_Lipidomics_results_all.tsv', sep="\t")
df
```

```python
df_D0 = df[["feature"] + df.filter(regex="D0$").columns.tolist()]
column_order = [
    'BS-064', 'JV-035', 'JV-048', 'KT-515', 'KT-926', 'JV-148', 'JV-157',
    'KT-193', 'KT-247', 'KT-312', 'KT-347', 'KT-352', 'KT-417', 'KT-434',
    'KT-522', 'KT-525', 'KT-565', 'KT-612', 'KT-705', 'KT-718', 'JV-071',
    'BS-082', 'BS-336', 'BS-346', 'BS-351', 'BS-364', 'BS-377', 'JV-138',
    'KT-313', 'KT-412', 'KT-445', 'KT-537', 'KT-538', 'KT-539', 'KT-663',
    'KT-695', 'KT-716', 'KT-723', 'KT-771', 'KT-805', 'KT-880', 'KT-974',
    'BS-671'
]
# Ajoute le suffixe D0 à chaque élément de la liste
column_order_d0 = [col + 'D0' for col in column_order]
conditions_column = df_D0.pop('feature')
df_D0 = df_D0.reindex(column_order_d0, axis=1)
if conditions_column is not None:
    df_D0.insert(0, 'feature', conditions_column)
df_D0.fillna(0, inplace=True)
df_D0 = df_D0.replace(',', '.', regex=True)
df_D0
```

```python
duplicates = df_D0[df_D0.duplicated(subset=['feature'], keep='first')]
duplicates
```

```python
df_D0.to_csv('/Users/loictalignani/research/project/lipidomics/data/lipidsig/Lipid_abundance_data_D0.tsv', sep="\t", index=False)
```
