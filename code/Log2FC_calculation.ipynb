# Calculate Log2 Fold-Change
Normalize dataset

```python
import pandas as pd
import numpy as np
```

```python
# Charger le fichier CSV
df = pd.read_csv(
    "/Users/loictalignani/research/project/lipidomics/data/pretreatment/MISSE_Global_Lipidomics_results_all_for_log2FC_calculation.tsv", sep="\t")
df
```

```python
df.dtypes
```

```python
# Extraire les noms de colonnes correspondant aux échantillons
sample_columns = [
    col for col in df.columns if "D0" in col or "D03" in col or "D10" in col or "D60" in col]
```

```python
# Créer de nouvelles colonnes avec le log2 fold-change
for col in sample_columns:
    patient_id = col[:-2]  # Retirer le suffixe temporel (D0, D03, etc.)
    if f"{patient_id}D60" in df.columns:  # Vérifier que D60 existe pour ce patient
        df[f"log2FC_D0-{patient_id}"] = np.log2(
            (df[f"{patient_id}D0"] + 1) / (df[f"{patient_id}D60"] + 1))
        df[f"log2FC_D03-{patient_id}"] = np.log2(
            (df[f"{patient_id}D03"] + 1) / (df[f"{patient_id}D60"] + 1))
        df[f"log2FC_D10-{patient_id}"] = np.log2(
            (df[f"{patient_id}D10"] + 1) / (df[f"{patient_id}D60"] + 1))
```
