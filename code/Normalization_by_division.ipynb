# Normalization by simple ratio
Normalize dataset

```python
import pandas as pd
import numpy as np
from scipy.stats import rankdata
```

```python
# Charger le fichier CSV
df = pd.read_csv(
    "/Users/loictalignani/research/project/lipidomics/data/pretreatment/MISSE_Global_Lipidomics_results_all_for_log2FC_calculation.tsv", sep="\t")
df
```

```python
offset = 1  # Offset à ajouter pour éviter la division par zéro
```

```python
# Identifier les patients en récupérant les préfixes des colonnes
patients = set(col.split('D')[0]
               for col in df.columns[1:])  # Exclure 'feature'
```

```python
# Appliquer l'offset à toutes les colonnes de patients
for patient in patients:
    patient_cols = [col for col in df.columns if col.startswith(patient)]
    # Ajout de l'offset pour toutes les colonnes du patient
    df[patient_cols] += offset
```

```python
# Calculer les ratios et créer les nouvelles colonnes
for patient in patients:
    d0_col, d3_col, d10_col, d60_col = f"{patient}D0", f"{patient}D03", f"{patient}D10", f"{patient}D60"

    if d60_col in df.columns:
        df[f"{patient}_D0_D60"] = df[d0_col] / df[d60_col]
        df[f"{patient}_D3_D60"] = df[d3_col] / df[d60_col]
        df[f"{patient}_D10_D60"] = df[d10_col] / df[d60_col]
```

```python
# Extraire les nouvelles colonnes dans trois DataFrames distincts
columns_D0_D60 = ["feature"] + \
    [col for col in df.columns if col.endswith("_D0_D60")]
columns_D3_D60 = ["feature"] + \
    [col for col in df.columns if col.endswith("_D3_D60")]
columns_D10_D60 = ["feature"] + \
    [col for col in df.columns if col.endswith("_D10_D60")]
```
